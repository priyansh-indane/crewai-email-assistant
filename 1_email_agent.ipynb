```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import LLM

llm=LLM(
    model="gemini/gemini-2.5-flash",
    temperature=0.1
)
llm.call("Hello !!")
```

```python
from crewai import Agent , Task , Crew

email_assistant= Agent(
    role="email assistant agent",
    goal="improve emails and make them sound clear and professional",
    backstory=" highly experienced skilled in writing email.",
    verbose=True,
    llm=llm


)

original_email=(" "
                "hello just wanted to tell that there is still some work left iam working on it thank you.")

email_task=Task(
    description=f"Take the following email and rewrite it into a professional version{original_email}",
    agent=email_assistant,
    expected_output="A professional written mail with proper formatting"
)

crew=Crew(
    agents=[email_assistant],
    tasks=[email_task],
    verbose=True
)

result=crew.kickoff()
print(result)
```
